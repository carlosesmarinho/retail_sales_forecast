<span style="font-size: 18px;">Dates and sales - Historical data</span>


__`Step 06`__ Analyze historic values for sales.

```python
df_dates = pd.to_datetime(df_train["date"])
print(f'primeira data: {df_dates.min()}')
print(f'última data: {df_dates.max()}')
```

```python
import matplotlib.pyplot as plt

# Convert the "date" column to datetime 
df_train['date'] = pd.to_datetime(df_train['date'])

# Group by date to sum daily sales
df_sales_by_date = df_train.groupby('date')[['sales']].sum()

# Create the figure and axis
fig, ax = plt.subplots(figsize=(16, 8))

# Plot the data
ax.plot(df_sales_by_date.index, df_sales_by_date['sales'], color='blue', linewidth=2)

# Customize the chart
ax.set_title('Vendas ao Longo do Tempo', fontsize=16)
ax.set_xlabel('Data', fontsize=12)
ax.set_ylabel('Vendas Totais', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Adjust layout and show the plot
plt.tight_layout()
plt.show() 
```

##### **Conclusão**: O gráfico mostra as vendas totais ao longo do tempo, com tendência de crescimento e alta variabilidade diária. Observam-se quedas abruptas em 2015 e 2017, que podem estar relacionadas a eventos externos ou anomalias nos dados. Esses padrões destacam a importância de tratar outliers e considerar sazonalidades na modelagem.

<span style="font-size: 18px;">Product Category Analysis</span>


__`Step 07`__ Analyze product category and its correlation with sales

##### Top category (family) by sales

```python
# Calculate total sales per product family
df_sales_family = df_train.groupby(['family'])[['sales']].sum().sort_values(by = 'sales', ascending = False)

# Calculate the percentage of sales per family
df_sales_family['percent'] = (df_sales_family['sales'] / df_sales_family['sales'].sum()) * 100
df_sales_family["cumulative_percent"] = df_sales_family["percent"].cumsum()

# Reset the index
df_sales_family.reset_index(inplace=True)

# Rename the columns
df_sales_family.rename(columns={"family": "#sales"}, inplace=True)
```

```python
# Select the top 15 product families
top_15 = df_sales_family.head(15).copy()

# Recalculate percentage of sales among the top 15
top_15.loc[:, 'percent_sales'] = (top_15['sales'] / top_15['sales'].sum()) * 100

# Create a figure with two subplots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Bar plot (sales percentage by category)
bars = ax1.barh(top_15['#sales'], top_15['percent_sales'], color='skyblue')
ax1.set_title('Famílias com maior volume de venda (95% do Total)', fontsize=16)
ax1.set_xlabel('Percentual de Vendas (%)', fontsize=12)
ax1.set_ylabel('Família de Produtos', fontsize=12)

# Add value labels on bars
for bar in bars:
    width = bar.get_width()
    ax1.text(width, bar.get_y() + bar.get_height()/2, f'{width:.1f}%', 
             ha='left', va='center', fontweight='bold', fontsize=10)

ax1.invert_yaxis()
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.xaxis.grid(True, linestyle='--', alpha=0.7)

# Line plot (cumulative percentage)
ax2.plot(top_15['#sales'], top_15['cumulative_percent'], marker='o', color='red', linewidth=2)
ax2.set_title('Percentual Cumulativo de Vendas', fontsize=16)
ax2.set_xlabel('Família de Produtos', fontsize=12)
ax2.set_ylabel('Percentual Cumulativo (%)', fontsize=12)
ax2.set_ylim(0, 100)
ax2.yaxis.grid(True, linestyle='--', alpha=0.7)

# Add labels to the line plot
for i, txt in enumerate(top_15['cumulative_percent']):
    ax2.annotate(f'{txt:.1f}%', (i, txt), textcoords="offset points", xytext=(0,10), 
                 ha='center', fontsize=10, fontweight='bold')

# Rotate x-axis labels
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha='right')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

##### **On promotion analyisis** #####

__`Step 08`__ Analyze product promotion and its correlation with sales

```python
import matplotlib.pyplot as plt

# Group by date to sum total promotions per day
df_sales_by_date = df_train.groupby('date')[['onpromotion']].sum()

# Create the figure and axis
fig, ax = plt.subplots(figsize=(16, 8))

# Plot the data
ax.plot(df_sales_by_date.index, df_sales_by_date['onpromotion'], color='blue', linewidth=2)

# Customize the plot
ax.set_title('Promos ao Longo do Tempo', fontsize=16)
ax.set_xlabel('Data', fontsize=12)
ax.set_ylabel('Promos Totais', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# Rotate x-axis labels
plt.xticks(rotation=45)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Group promotions by product family
df_onpromotion_family = df_train.groupby('family')[['onpromotion']].sum().sort_values(by='onpromotion', ascending=False)

# Calculate percentage and cumulative percentage
df_onpromotion_family['percent'] = (df_onpromotion_family['onpromotion'] / df_onpromotion_family['onpromotion'].sum()) * 100
df_onpromotion_family['cumulative_percent'] = df_onpromotion_family['percent'].cumsum()

# Reset index
df_onpromotion_family.reset_index(inplace=True)
```

```python
# Select only families that together account for up to 95% of promotions
df_filtered = df_onpromotion_family[df_onpromotion_family['cumulative_percent'] <= 95].copy()
```

```python
# Recalculate percentages just within the subset (for a more accurate plot)
df_filtered['percent_onpromotion'] = (df_filtered['onpromotion'] / df_filtered['onpromotion'].sum()) * 100

# Criar figura com dois subplots lado a lado
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# Gráfico de barras
bars = ax1.barh(df_filtered['family'], df_filtered['percent_onpromotion'], color='#FFA07A')
ax1.set_title('Famílias com Mais Promoções (95% do Total)', fontsize=16)
ax1.set_xlabel('Percentual de Promoções (%)', fontsize=12)
ax1.set_ylabel('Família de Produtos', fontsize=12)
for bar in bars:
    width = bar.get_width()
    ax1.text(width, bar.get_y() + bar.get_height()/2, f'{width:.1f}%', 
             ha='left', va='center', fontweight='bold', fontsize=10)
ax1.invert_yaxis()
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.xaxis.grid(True, linestyle='--', alpha=0.7)

# Gráfico de linha
ax2.plot(df_filtered['family'], df_filtered['cumulative_percent'], marker='o', color='#FF6347', linewidth=2)
ax2.set_title('Percentual Cumulativo de Promoções', fontsize=16)
ax2.set_xlabel('Família de Produtos', fontsize=12)
ax2.set_ylabel('Percentual Cumulativo (%)', fontsize=12)
ax2.set_ylim(0, 100)
ax2.yaxis.grid(True, linestyle='--', alpha=0.7)
for i, txt in enumerate(df_filtered['cumulative_percent']):
    ax2.annotate(f'{txt:.1f}%', (i, txt), textcoords="offset points", xytext=(0,10), 
                 ha='center', fontsize=10, fontweight='bold')
plt.setp(ax2.xaxis.get_majorticklabels(), rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# 2. Converter datas para datetime
df_train['date'] = pd.to_datetime(df_train['date'])
df_holidays['date'] = pd.to_datetime(df_holidays['date'])
df_oil_price['date'] = pd.to_datetime(df_oil_price['date'])
df_transaction['date'] = pd.to_datetime(df_transaction['date'])

# 3. Pré-processar feriados por localidade
## Excluir types que não nos interessam:
df_holidays = df_holidays[(df_holidays['type'] != 'Work Day') | (df_holidays['type'] != 'Events')]

## 3.1. definir categorias de feriados:
df_holidays['state'] = np.where(df_holidays['locale'] == 'Regional', df_holidays['locale_name'], np.nan)
df_holidays['city'] = np.where(df_holidays['locale'] == 'Local', df_holidays['locale_name'], np.nan)
df_holidays['holiday_state'] = np.where((df_holidays['locale'] == 'Regional') & (df_holidays['transferred'] == False), 1, 0)
df_holidays['holiday_city'] = np.where((df_holidays['locale'] == 'Local') & (df_holidays['transferred'] == False), 1, 0)
df_holidays['holiday_national'] = np.where((df_holidays['locale'] == 'National') & (df_holidays['transferred'] == False), 1, 0)

## 3.2. Criação de df por categoria de feriado
df_holidays_state = df_holidays[df_holidays['holiday_state'] == 1]
df_holidays_state = df_holidays_state[['date','state', 'holiday_state']].drop_duplicates()

df_holidays_city = df_holidays[df_holidays['holiday_city'] == 1]
df_holidays_city = df_holidays_city[['date','city', 'holiday_city']].drop_duplicates()

df_holidays_national = df_holidays[df_holidays['holiday_national'] == 1]
df_holidays_national = df_holidays_national[['date','holiday_national']].drop_duplicates()


# 4. Merge com lojas e feriados (base principal é df_train)
df_merged = df_train.merge(df_stores, on='store_nbr', how='left')
df_merged = df_merged.merge(df_holidays_city.drop_duplicates(), on=['date', 'city'], how='left')
df_merged = df_merged.merge(df_holidays_state.drop_duplicates(), on=['date', 'state'], how='left')
df_merged = df_merged.merge(df_holidays_national.drop_duplicates(), on='date', how='left')
df_merged['holiday'] = df_merged[['holiday_city', 'holiday_state','holiday_national']].max(axis=1)
display(df_merged)
df_merged =  df_merged.drop(columns = ['holiday_city', 'holiday_state','holiday_national'])
df_merged['holiday'] = np.where(df_merged['holiday'].isnull(),0,df_merged['holiday'])

# 5. Merge com transactions
df_merged = df_merged.merge(
    df_transaction[['date', 'store_nbr', 'transactions']],
    on=['date', 'store_nbr'], how='left'
)

df_merged['transactions'] = df_merged['transactions'].fillna(0)


# 6. Merge com preço do petróleo
df_merged = df_merged.merge(df_oil_price, on='date', how='left')
df_merged["dcoilwtico"] = df_merged["dcoilwtico"].interpolate(method='linear').bfill().ffill()

# 7. Ajustes de tipos
df_merged['cluster'] = df_merged['cluster'].astype(str)
df_merged['onpromotion'] = df_merged['onpromotion'].astype(float)
df_merged['holiday'] = df_merged['holiday'].astype(str)

# 8. Sanity check de integridade
assert df_merged.shape[0] == df_train.shape[0], "🚨 Algo deu errado: número de linhas diferente do df_train!"
print("✅ Todas as linhas do df_train estão presentes no df_merged.")

# 9. Sanity check de shape e preview
print("\n📐 Shape final do DataFrame:", df_merged.shape)
print("\n🔍 Visualização inicial:")
display(df_merged.head())

# 10. Salvar em Parquet
df_merged.to_parquet('df_merged.parquet')
```
